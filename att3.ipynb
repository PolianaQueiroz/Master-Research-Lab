```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
import seaborn as sns

#gerando gráfico de dados aleatórios
x, y = make_regression(n_samples = 150, n_features=1, noise=30)
plt.scatter(x, y)
```

```python
#treinando o mmodelo
linear_regression = LinearRegression()
linear_regression.fit(x,y)
```

```python
#calculndo coeficiente amgulae e linear
coeficiente_angular = linear_regression.coef_
coeficiente_linear = linear_regression.intercept_

print(f"Coeficiente angular = {coeficiente_angular}")
print(f"Coeficiente linear = {coeficiente_linear}")
```

```python
# Mostrando função linear no gráfico
plt.scatter(x, y)
plt.plot(x, coeficiente_linear + coeficiente_angular*x, color='red')
plt.show()
```

```python
plt.scatter(x, y)
plt.plot(x, coeficiente_linear + coeficiente_angular*x, color='red')
plt.scatter(1.5, coeficiente_linear + coeficiente_angular*1.5, color='black', s = 150)
plt.show()
```
